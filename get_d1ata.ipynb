```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
# for i in range(2,4):
#     r = requests.get(f'https://angelmatch-store.sfo3.digitaloceanspaces.com/pitch-decks/pdf/{i}.pdf')
#     with open(f'{i}.pdf', 'wb') as f:
#         f.write(r.content)
```

```python

names=[]
descr=[]
funds=[]
tag=[]
recent_funding=[]
stage=[]
pdf=[]
year=[]
for i in range(34):
    req=requests.get(f'https://angelmatch.io/pitch_decks?page={i}')
    soup = BeautifulSoup(req.content, 'html.parser')

    # with open('test.txt', 'w') as f:
    #     f.write(str(soup))
    div = soup.select('main div')
    js = soup.select('script')[19:]
    
    
    for j in js:
        # print(j.text)
        fi = json.loads(str(j.text))
        for startup in fi['props']['pageProps']['list']['content']:
            pdf.append(startup['pdf'])
            names.append(startup['name'])
            tag.append(startup['tagline'])
            descr.append(startup['description'])
            recent_funding.append(startup['raised'])
            funds.append(startup['totalFundingAmount'])
            stage.append(startup['stages'][0]['title'])
            year.append(startup['year'])
```

```python
import pandas as pd
df = pd.DataFrame([names,tag,funds,recent_funding,year,stage,pdf]).T
df.columns = ['Name','Tagline','Funds','Recent_funding','year','stage','pdf']
df.to_parquet('startups.parquet')
```

```python
# startups['labels'] 
startups['labels'] = pd.qcut(startups['Num'].values.round(),10)
startups=startups.rename(columns={'PDFtext':'text'})
pd.qcut(startups['Num'].values.round(),10).categories
```

```python
bertinp = startups[['text','labels']]
tdf = bertinp.iloc[:570,:]
vdf = bertinp.iloc[570:,:]
```

```python
vdf.labels.value_counts()
```

```python
tdf=tdf[tdf.text.str.strip() != '']
vdf=vdf[vdf.text.str.strip() != '']
```

```python
def remove_non_ascii_2(string):
    return string.encode('ascii', errors='ignore').decode()
tdf['text']=tdf.text.apply(remove_non_ascii_2)
vdf['text']=vdf.text.apply(remove_non_ascii_2)
```

```python
tdf.reset_index(drop=True,inplace=True)
vdf.reset_index(drop=True,inplace=True)
tdf.labels = pd.Categorical(tdf.labels)
vdf.labels = pd.Categorical(vdf.labels)
tdf['labels'] = tdf.labels.cat.codes
vdf['labels'] = vdf.labels.cat.codes
```

```python
tdf=pd.DataFrame(tdf.to_dict())
vdf=pd.DataFrame(vdf.to_dict())
```
